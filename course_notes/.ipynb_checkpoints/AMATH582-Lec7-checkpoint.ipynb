Demo code for lecture 7. Demonstrating how scipy and PyWavelets can be used for image processing. 


```python
import numpy as np
import scipy as sp
import pywt
import matplotlib.pyplot as plt
```

```python
from scipy import misc


# import a test image
face = misc.face(gray=True)
face = face[:512, -512:]  # crop out square on right



# make a noisy image
noisy_face = np.copy(face).astype(np.float)
noisy_face += face.std() * 0.75 * np.random.standard_normal(face.shape)


fig, ax = plt.subplots(figsize=(10,10))


ax.imshow(noisy_face, cmap=plt.cm.gray)
ax.axis('off')


plt.show()
```

Our ultimate goal is to denoise this image by filtering its Fourier transform. But our first task is to compute its fft. 

```python
face_hat = np.fft.fft2(face)
face_hat = np.fft.fftshift(face_hat) # we still need to shift the fft in 2D as well 

noisy_face_hat = np.fft.fft2(noisy_face)
noisy_face_hat = np.fft.fftshift(noisy_face_hat) # we still need to shift the fft in 2D as well 


k_grid = np.linspace(-256, 256, 512)

k1, k2 = np.meshgrid(k_grid, k_grid)

from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0][0].imshow(face, cmap=plt.cm.gray)
ax[0][0].axis('off')
ax[0][0].set_title('Original')

cont_levels = np.arange(0, 20, 0.5)

im1 = ax[0][1].contourf(k1, k2, np.log(np.abs(face_hat)), cont_levels)
ax[0][1].set_xlim(-200,+200)
ax[0][1].set_ylim(-200,+200)

divider = make_axes_locatable(ax[0][1])
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im1, cax=cax)

ax[0][1].set_xlabel('kx')
ax[0][1].set_ylabel('ky')

ax[1][0].imshow(noisy_face, cmap=plt.cm.gray)
ax[1][0].axis('off')
ax[1][0].set_title('Noisy')



im2 = ax[1][1].contourf(k1, k2, np.log(np.abs(noisy_face_hat)), cont_levels)
ax[1][1].set_xlim(-200,+200)
ax[1][1].set_ylim(-200,+200)



divider = make_axes_locatable(ax[1][1])
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im2, cax=cax)

ax[1][1].set_xlabel('kx')
ax[1][1].set_ylabel('ky')


plt.show()
```

Exactly like the 1D case you can manipulate the 2D fft and try to reconstruct the image with some filtering. 

```python
def g(x,y, s): # 2D Gaussian filter
  val = np.exp( - ( (x**2 + y**2)/(2*s**2)  ))

  return val
```

```python
# get filter values in freq domain and plot it
sigma = 50 # kill stuff outside k = -100 and 100

g_vals = g(k1, k2, sigma)

fig, ax = plt.subplots(figsize=(5,5))

ax.contourf(k1, k2, g_vals)
ax.set_title('Gaussian filter')
ax.set_xlabel('kx')
ax.set_xlabel('ky')
```

```python
# multiply fft of noisy face with filter and reconstruct 

noisy_face_hat_filtered = g_vals*noisy_face_hat

noisy_face_filtered = np.real(np.fft.ifft2( np.fft.ifftshift(noisy_face_hat_filtered) ))
```

```python
fig, ax = plt.subplots(3,2, figsize=(10,15))

ax[0][0].imshow(face, cmap=plt.cm.gray)
ax[0][0].axis('off')
ax[0][0].set_title('Original')

cont_levels = np.arange(0, 20, 0.5)

im1 = ax[0][1].contourf(k1, k2, np.log(np.abs(face_hat)), cont_levels)
ax[0][1].set_xlim(-200,+200)
ax[0][1].set_ylim(-200,+200)

divider = make_axes_locatable(ax[0][1])
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im1, cax=cax)

ax[0][1].set_xlabel('kx')
ax[0][1].set_ylabel('ky')

ax[1][0].imshow(noisy_face, cmap=plt.cm.gray)
ax[1][0].axis('off')
ax[1][0].set_title('Noisy')



im2 = ax[1][1].contourf(k1, k2, np.log(np.abs(noisy_face_hat)), cont_levels)
ax[1][1].set_xlim(-200,+200)
ax[1][1].set_ylim(-200,+200)



divider = make_axes_locatable(ax[1][1])
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im2, cax=cax)

ax[1][1].set_xlabel('kx')
ax[1][1].set_ylabel('ky')


ax[2][0].imshow(noisy_face_filtered, cmap=plt.cm.gray)
ax[2][0].axis('off')
ax[2][0].set_title('Denoised')



im3 = ax[2][1].contourf(k1, k2, np.log(np.abs(noisy_face_hat_filtered)), cont_levels)
ax[2][1].set_xlim(-200,+200)
ax[2][1].set_ylim(-200,+200)



divider = make_axes_locatable(ax[2][1])
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im3, cax=cax)

ax[2][1].set_xlabel('kx')
ax[2][1].set_ylabel('ky')


plt.show()
```

We now apply the discrete wavelet transform to the same image to extract its features. 

```python
face_A, (face_DH, face_DV, face_DD) = pywt.dwt2( face, 'haar' )

fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0][0].imshow(face_A, cmap=plt.cm.gray)
ax[0][0].axis('off')
ax[0][0].set_title('Coarse approximation')

ax[0][1].imshow(np.abs(face_DH), cmap=plt.cm.gray)
ax[0][1].axis('off')
ax[0][1].set_title('|Horizontal DWT Coeff|')

ax[1][0].imshow(np.abs(face_DV), cmap=plt.cm.gray)
ax[1][0].axis('off')
ax[1][0].set_title('|Vertical DWT Coeff|')

ax[1][1].imshow(np.abs(face_DD), cmap=plt.cm.gray)
ax[1][1].axis('off')
ax[1][1].set_title('|Diagonal DWT Coeff|')
```
