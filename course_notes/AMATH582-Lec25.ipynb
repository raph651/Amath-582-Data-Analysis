```python
import cvxpy as cvx
```

```python
# synthetic signal to be reconstructed 

def f(t): 

  val = 0.6*np.exp( - np.abs( t - 0.15)**2/(2*0.1**2) )*np.cos(np.pi*12*t)\
  + (t>= 0.5)*(t<=0.8)

  return val 

N = 128

t = np.linspace(0, 1, N)

f_vals = f(t)


plt.plot(t, f_vals)
plt.xlabel('t')
plt.ylabel('g(t)')
plt.title('Original function')
```

We will observe a noisy subsample of the signal. 

```python
# extract small sample of signal
M = int(0.4*N) # 30% sample
np.random.seed(seed=10)

S = np.random.randn(M, N)

#indx = np.random.choice(N, M, replace=False) # random sample of indices
#indx.sort() # sorting not strictly necessary, but convenient for plotting
#t_sub = t[indx]
#y = f_vals[indx]

y = np.dot(S, f_vals)


fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(t, f_vals)
ax[0].set_xlabel('t')

ax[0].set_title('Original function f')

ax[1].scatter(np.arange(M), y, s= 30, c='r')
ax[1].set_title('Measrements y')

#plt.title('Original function and subsample')
```

```python
# compute wavelet transform of the original signal 

wlt = 'haar'

wg_dict = pywt.wavedec( f_vals, wavelet = wlt )

wg, wg_slices = pywt.coeffs_to_array(wg_dict)

NW = len(wg)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.bar(np.arange(len(wg)), np.abs(wg))
ax.set_xlabel('index j')
ax.set_title('|Wavelet coeffs.| of signal')
```

We now setup an optimization problem to reconstruct the original function $f$ from the sparse observations in f_sub.  

```python
# we need to construct the matrix that reconstrauct a function from its 
# vector of wavelet coefficients

W = np.zeros( (NW, NW) )

I = np.eye(NW)

for i in range(N):

  W[:, i] = pywt.waverec( pywt.array_to_coeffs(I[:,i], wg_slices, output_format='wavedec') , wavelet = wlt )

# columns of A are nothing but the wavelet basis 

for i in range(10): 

  plt.plot(t, W[:,i])
```

```python
plt.plot(np.dot(W, wg))
```

```python
# we also need the subsampling matrix, ie, the matrix that extracts the values of a function at the random 
# observation points that were used to obtain the data y

#S = I[indx, :] # this is nothing but a sub matrix of the identity consisting of the rows that correspond to the observation points 

#S = I[indx, :]

A = np.dot(S, W) # A = S*W  
```

```python
# we now setup two optimization problems using cvxpy, to find the wavelet coeffs of the function f. 

# first formulation using l2 penalty

x_l2 = cvx.Variable(N)
tol = 5e-4

objective_l2 = cvx.Minimize( 0.1*cvx.sum_squares( x_l2) + cvx.sum_squares( A @ x_l2 - y ) )

# alt formulation
# objective_l2 = cvx.Minimize( 0.1*cvx.sum_squares( x_l2) )
#constraints_l2 = [-tol <= A*x_l2 - y, A*x_l2 - y <= tol]
#prob_l2 = cvx.Problem( objective_l2, constraints_l2 )

prob_l2 = cvx.Problem( objective_l2)
prob_l2.solve()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10, 10))

ax[0].bar(np.arange(len(wg)), np.abs(wg))
ax[0].set_xlabel('index j')
ax[0].set_title('|Wavelet coeffs.| of signal')

ax[1].bar(np.arange(len(wg)), np.abs(x_l2.value))
ax[1].set_xlabel('index j')
ax[1].set_title('$l_2$ reconstruction')
```

```python
# plot original function against l2 reconstruction 

plt.plot(t, f_vals, 'r')
plt.plot(t, np.matmul(W, np.transpose(x_l2.value)), 'k')
plt.xlabel('t')
plt.title('Original function vs l_2 reconstruction')
```

```python
# second formulation using l1 penalty

x_l1 = cvx.Variable(N)

objective_l1 = cvx.Minimize( 0.1*cvx.norm( x_l1, 1 ) + cvx.sum_squares( A @ x_l1 - y) )
prob_l1 = cvx.Problem( objective_l1)

# alt formulation
#objective_l1 = cvx.Minimize( cvx.norm( x_l1, 1 ) )
#constraints_l1 =  [-tol <= A*x_l1 - y, A*x_l1 - y <= tol]
#prob_l1 = cvx.Problem( objective_l1, constraints_l1)

prob_l1.solve()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10, 10))

ax[0].bar(np.arange(len(wg)), np.abs(wg))
ax[0].set_xlabel('index j')
ax[0].set_title('|Wavelet coeffs.| of signal')

ax[1].bar(np.arange(len(wg)), np.abs(x_l1.value))
ax[1].set_xlabel('index j')
ax[1].set_title('$l_1$ reconstruction')
```

```python
# plot original function against l1 reconstruction 

plt.plot(t, f_vals, 'b')
plt.plot(t, np.dot(W, x_l1.value), 'k')
plt.xlabel('t')
plt.title('Original function vs l_1 reconstruction')
```

Another example with a signal that is sparse in the cosine basis. 

```python
import scipy.fftpack as spfft
```

```python
NN = 5000

tt = np.linspace(0, 1/8, NN)

f = np.sin(1394 * np.pi * tt) + np.sin(3266 * np.pi * tt) # original signal, just a sum of two sinosoids

ft = spfft.dct( f, norm = 'ortho' ) # compute discrete cosine transform of f 
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))

ax.plot(np.abs(ft))
ax.set_xlabel('l - freq. index')
ax.set_ylabel('|dct(f)|')
```

Clearly the signal has a very sparse cosine transform. 

```python
# we only observe a random subsample of the signal. 

MM = 500 # only 10% of values are observed 

indx = np.random.choice(NN, MM, replace=False)
indx.sort()
ttt = tt[indx]
yy = f[indx]
```

```python
# plot original signal f and data y 

fig, ax = plt.subplots(1,2, figsize=(20,5))

ax[0].plot(tt, f)
ax[0].scatter(ttt, yy, s=30, c='r' )
ax[0].set_xlabel('t')
ax[0].set_ylabel('f')

ax[0].set_title('f and y')


ax[1].plot(tt, f)
ax[1].scatter(ttt, yy, s=30, c='r' )
ax[1].set_xlim([0, 0.02])
ax[1].set_xlabel('t')
ax[1].set_ylabel('f')

ax[1].set_title('f and y (zoom)')
```

Our goal is to recover $f$ (the blue signal) from $y$ (the red dots). 

```python
# construct A matrix that goes from dct coefficients to sparse measurements y

I = np.eye(NN)
S = I[indx, :] # subsampled identity matrix 

D = spfft.idct(np.eye(NN), norm = 'ortho', axis=0)

A = np.dot(S,D)
```

```python
## sanity check that A does the right thing 

print(np.linalg.norm( yy - np.dot(A, ft ) )) # should be very small or zero
```

```python
# solve l1 regularized problem using cvx 

vb = cvx.Variable(NN) #vb for vector beta
objective_dct = cvx.Minimize( cvx.norm(vb, 1) )
constraints_dct = [A*vb == yy] # equality constraints at the observation points

problem_dct = cvx.Problem(objective_dct, constraints_dct)
```

```python
result = problem_dct.solve(verbose=True)
```

```python
# plot original f and the reconstruction

f_rec = spfft.idct(vb.value, norm='ortho', axis = 0)

# plot original signal f and data y 

fig, ax = plt.subplots(2,2, figsize=(20,5))

ax[0,0].plot(tt, f)
ax[0,0].set_xlabel('t')
ax[0,0].set_ylabel('f')
ax[0,0].set_title('Original signal')


ax[0,1].plot(tt, f)
ax[0,1].set_xlim([0, 0.02])
ax[0,1].set_xlabel('t')
ax[0,1].set_ylabel('f')
ax[0,1].set_title('Original signal zoom')

ax[1,0].plot(tt, f_rec)
ax[1,0].set_xlabel('t')
ax[1,0].set_ylabel('f')
ax[1,0].set_title('Reconstructed signal')


ax[1,1].plot(tt, f_rec)
ax[1,1].set_xlim([0, 0.02])
ax[1,1].set_xlabel('t')
ax[1,1].set_ylabel('f')
ax[1,1].set_title('Reconstructed signal zoom')


```
