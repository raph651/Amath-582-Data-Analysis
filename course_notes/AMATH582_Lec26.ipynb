```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(t): 

#  return np.maximum(0, t) # relu

#  return np.maximum(0, t) + 0.01*t # Leaky Relu

#  return 1/( 1 + np.exp(-t)) # sigmoid

  return np.cos(t) # cosine

#  return np.maximum(0, 1 - np.abs(t)) # hat

# 
```

```python

np.random.seed(42)

L = 5

ss = 1

W0 = ss*np.random.randn(3) # input W matrix

W = ss*np.random.randn(L, 3, 3) # hidden W matrices

WL = ss*np.random.rand(3) # output W matrix

b = ss*np.random.randn(L+1,3) # input + hidden biases

bL = 0*np.random.randn(1) # output bias
```

```python
def FNN(x):

  z = W0*g(x) + b[0]

  for i in range(L):

    #print('L : ', i)
    #print(np.dot(W[i], z) + b[i+1])
    #print(g(np.dot(W[i], z)) + b[i+1])
    

    z = np.dot(W[i], g(z)) + b[i+1]

  val = np.dot(WL, z) + bL

  return val
```

```python
FNNv = np.vectorize(FNN)
```

```python
Nx = 100
xx = np.linspace(0,2*np.pi, Nx)
FNNvals = np.zeros(Nx)

for i in range(Nx):

  FNNvals[i] = FNN(xx[i])
```

```python
plt.plot(xx, FNNvals)
```
