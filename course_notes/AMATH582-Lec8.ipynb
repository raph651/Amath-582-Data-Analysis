Notebook to go along with Lecture 8. Here we see how to add noise to images in Python and then how to denoise/filter images using DFT and scipy.ndimage. Afterwards we do image compression using DWT.

```python
import numpy as np
import pywt
import matplotlib.pyplot as plt
from scipy import misc
from scipy import ndimage
from scipy import signal
```

```python


# import a test image
face = misc.face(gray=True)
face = face[:512, -512:]  # crop out square on right



# make a noisy image by adding Gaussian noise 

def add_Gaussian_noise( img, snr):

  # img : original image
  # snr : signal to noise ratio, i.e., ratio of signal standard deviation to that of noise

  val = np.copy(img).astype(np.float)

  val += face.std() * (1/snr) * np.random.standard_normal(face.shape)

  return val



noisy_face_Gaussian = add_Gaussian_noise(face, (1/0.5) )

fig, ax = plt.subplots(1,2, figsize=(10,20))

ax[0].imshow(face, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(noisy_face_Gaussian, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Noisy (Gaussian)')


plt.show()
```

```python
# alternative to the Gaussian noise we can add salt and pepper 

def add_sp_noise(img, prob):
    
    # img : original image
    # prob: probability of noise occuring at a pixel

    val = np.copy(img).astype(np.float)
    
    probs = np.random.random(val.shape[:2])
    val[probs < (prob / 2)] = 0
    val[probs > 1 - (prob / 2)] = 255
    return val

noisy_face_sp = add_sp_noise(face, 0.1 )

fig, ax = plt.subplots(1,2, figsize=(10,20))

ax[0].imshow(face, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(noisy_face_sp, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Noisy (Salt and Pepper)')


plt.show()
```

We now use filtering to denoise the s&p image. 

```python
from matplotlib.rcsetup import validate_bool_maybe_none
# define your filter (function to be multiplied in frequency domain)

def g_filter(x,y, s): # 2D Gaussian filter
  val = np.exp( - ( (x**2 + y**2)/(2*s**2)  ))

  return val

def filter_image( img, fltr, fltr_param):

  # img: original image
  # filter: filter function you defined earlier
  # filter_param: parameters to be passed to your filter

  img_hat = np.fft.fftshift( np.fft.fft2(img) )


  # in case img is not square
  N = img.shape
  k_grid_1 = np.linspace(-N[0]/2, N[0]/2, N[0])
  k_grid_2 = np.linspace(-N[1]/2, N[1]/2, N[1])

  k1, k2 = np.meshgrid(k_grid_1, k_grid_2)

  val = fltr(k1, k2, fltr_param)*img_hat

  return np.real(np.fft.ifft2( np.fft.ifftshift(val)))

filtered_face = filter_image(noisy_face_sp, g_filter, 10)

fig, ax = plt.subplots(1,3, figsize=(15,45))

ax[0].imshow(face, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(noisy_face_sp, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Noisy (Salt and Pepper)')

ax[2].imshow(filtered_face, cmap=plt.cm.gray)
ax[2].axis('off')
ax[2].set_title('Filtered')


plt.show()
```

Most likely you don't even need to do this much coding. Scipy.ndimage and Scipy.signal already have collections for standard filters for use. 

```python
blurred_face = ndimage.gaussian_filter(noisy_face_sp, sigma=3) # this convolves image with a Gaussian kernel in image domain. Fourier_filters under ndimage do the same as what we did above. 

median_face = ndimage.median_filter(noisy_face_sp, size=5)

wiener_face = signal.wiener(noisy_face_sp, (5, 5))


ig, ax = plt.subplots(1,5, figsize=(25,75))

ax[0].imshow(face, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(noisy_face_sp, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Noisy')

ax[2].imshow(blurred_face, cmap=plt.cm.gray)
ax[2].axis('off')
ax[2].set_title('Blurred')

ax[3].imshow(median_face, cmap=plt.cm.gray)
ax[3].axis('off')
ax[3].set_title('Median Filter')

ax[4].imshow(wiener_face, cmap=plt.cm.gray)
ax[4].axis('off')
ax[4].set_title('Wiener Filter')


plt.show()
```

We now turn to DWT and do an example in image compression. 

```python
#shape = x.shape

level = 1 # how many levels of DWT to perform

fig, axes = plt.subplots(figsize=[10, 10])

# compute the 2D DWT
c = pywt.wavedec2(face, 'haar', mode='periodization', level=level)

# normalize each coefficient array independently for better visibility
c[0] /= np.abs(c[0]).max()
for detail_level in range(level):
  c[detail_level + 1] = [d/np.abs(d).max() for d in c[detail_level + 1]]

# plot the normalized coefficients
arr, slices = pywt.coeffs_to_array(c)
axes.imshow(arr, cmap=plt.cm.gray)
axes.set_title('DWT \n({} levels)'.format(level))
axes.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# compute max level transform of the image and plot histogram of coefficients

face_dwt = pywt.wavedec2(face, 'haar', mode='periodization')

face_dwt_mat, face_dwt_slices = pywt.coeffs_to_array(face_dwt)


face_dwt_mat_normalized = face_dwt_mat/np.abs(face_dwt_mat).max()


#fig, axes = plt.subplots(figsize=[10, 10])
plt.hist(face_dwt_mat_normalized.flatten(), bins=2000)

plt.xlim(-0.01, 0.01)
plt.xlabel('Normalized value')
plt.ylabel('Number of coeffs.')
plt.show()
```

We will now compress the image by setting small coefficients to zero and keeping only a small number of larger coefficients.

```python
ratio = 0.2 # ratio of coefficients to keep (x100 to get percentages)

# flatten the dwt and sort it from largest to smallest coefficients in abs value 

sorted_face_dwt = np.sort( np.abs(face_dwt_mat).flatten() )

sorted_face_dwt = sorted_face_dwt[::-1]

# number of coefficients to keep above threshold
N_keep = int( ratio*len(sorted_face_dwt) )

Coeff_threshold = sorted_face_dwt[N_keep] # smallest abs coeff to keep will be our threshold

face_dwt_mat_compressed = pywt.threshold( face_dwt_mat, Coeff_threshold, mode='hard' )

# convert matrix back to pwt dictionary format and reconstruct
face_dwt_compressed  = pywt.array_to_coeffs( face_dwt_mat_compressed, face_dwt_slices, output_format='wavedec2' )

face_compressed = pywt.waverec2( face_dwt_compressed, wavelet='haar', mode= 'periodization' )

fig, ax = plt.subplots(1,2, figsize=(15,30))

ax[0].imshow(face, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(face_compressed, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Compressed (ratio '+ str(ratio) + ')')

plt.show()
```
